# Pitch (Field)

## Basic

```python
import matplotlib.pyplot as plt

from mplsoccer import Pitch
```

```python
pitch = Pitch()

fig, ax = pitch.draw()

plt.show()
```

## Title

```python
pitch = Pitch()

fig, ax = pitch.draw()

ax.set_title("Football Analytics", fontweight="bold")

plt.show()
```

## Data Providers

- `statsbomb`
- `opta`
- `tracab`
- `wyscout`
- `uefa`
- `metricasports`
- `custom`
- `skillcorner`
- `secondspectrum`
- `impect`

### StatsBomb

```python
pitch = Pitch(pitch_type="statsbomb")

fig, ax = pitch.draw()

plt.show()
```

### Metrica Sports

```python
pitch = Pitch(pitch_type="metricasports", pitch_length=105, pitch_width=68)

fig, ax = pitch.draw()

plt.show()
```

## Half Pitch

```python
pitch = Pitch(half=True)

fig, ax = pitch.draw()

plt.show()
```

## Vertical Pitch

```python
from mplsoccer import VerticalPitch
```

```python
pitch = VerticalPitch()

fig, ax = pitch.draw()

plt.show()
```

## Half Vertical Pitch

```python
pitch = VerticalPitch(half=True)

fig, ax = pitch.draw()

plt.show()
```

## Multiple Pitches

```python
pitch = Pitch()

fig, axs = pitch.draw(nrows=2, ncols=3)

plt.show()
```

### Line Color

```python
pitch = Pitch(line_color="black")

fig, axs = pitch.draw()

plt.show()
```

## Grass Field (잔디구장!)

```python
pitch = Pitch(
    pitch_color='grass',
    line_color='white',
    stripe=True
)

fig, ax = pitch.draw()

plt.show()
```

## Corner Arcs

```python
pitch = VerticalPitch(
    corner_arcs=True,
    line_color="black",
    half=True
)

fig, ax = pitch.draw()

plt.show()
```
