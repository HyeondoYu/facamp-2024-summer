## Rectangle

````python
from matplotlib.patches import Rectangle
````

### Filled

````python
fig, ax = plt.subplots()

# (x, y), width, height
my_rectangle = Rectangle((0.2, 0.2), 0.6, 0.4)
 
ax.add_patch(my_rectangle)

plt.show()
````

### Transparent

````python
fig, ax = plt.subplots()

my_rectangle = Rectangle((0.2, 0.2), 0.6, 0.4, fill=False)
 
ax.add_patch(my_rectangle)

plt.show()
````

## Circle

### Filled

````python
fig, ax = plt.subplots()
ax.set_aspect(1)

# (x, y), radius
my_circle = plt.Circle((0.5, 0.5), 0.2)
 
ax.add_patch(my_circle)

plt.show()
````

### Transparent

````python
fig, ax = plt.subplots()
ax.set_aspect(1)

my_circle = plt.Circle((0.5, 0.5), 0.2, fill=False)
 
ax.add_patch(my_circle)

plt.show()
````

## Arc

````python
from matplotlib.patches import Arc
````

The arc is drawn in the counterclockwise direction.

````python
fig, ax = plt.subplots()
ax.set_aspect(1)

my_arc = Arc(
    (0.5, 0.5),
    height=0.4, width=0.4,
    angle=0,
    theta1=0, theta2=180,
    color="red"
)

ax.add_patch(my_arc)
ax.grid()

plt.show()
````

`angle` sets a starting point.

````python
fig, ax = plt.subplots()
ax.set_aspect(1)

my_arc = Arc(
    (0.5, 0.5),
    height=0.4, width=0.4,
    angle=45,
    theta1=0, theta2=180,
    color="red"
)

ax.add_patch(my_arc)

ax.grid()

plt.show()
````
