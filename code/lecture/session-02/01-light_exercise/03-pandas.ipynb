## Sort

```python
df.sort_values(by="FIFAMarketValue", ascending=False)
```

## Group

Average wage per club

```python
df.groupby("ClubName").FIFAWage.mean()
```

## Group and Sort

```python
df.groupby("ClubName").FIFAWage.mean().sort_values(ascending=False)
```
